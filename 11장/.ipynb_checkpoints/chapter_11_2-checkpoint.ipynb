## 예제 11-2-1 help 내장 함수 알아보기

```python
help(int)                        ## help 클래스로 클래스의 문서화 정보를 출력한다
```

```python
def add_(x,y) :                                             ## 두 개의 매개변수를 가진 함수를 정의한다
                                                            ## 첫번째 문장에 문자열로 함수를 설명하는 문자열을 작성해서 문서화한다
    """ 이 함수는 두 개의 매개변수를 덧셈하는 함수입니다.
    """
    return x+y                                              ## 두 매개변수를 더한 값을 반환한다
```

```python
add_.__doc__                                                 ## 함수 문서화 속성 __doc__를 확인한다
```

```python
help(add_)                                                  ## 함수를 help로 문서화 정보를 확인한다 
                                                            ## 함수 이름과 문서화 정보를 같이 보여준다
```
